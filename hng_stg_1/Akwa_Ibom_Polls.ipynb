```python
import requests
import time

file_path = 'cleaned_data_no_duplicates.csv'
data = pd.read_csv(file_path)

# OpenCage API key
api_key = '83f25aa5c56e4d64b0ee5ee8658ae9d7'

# Extract unique polling unit names with missing coordinates
missing_coordinates = data[data['location.latitude'].isna() | data['location.longitude'].isna()]
unique_polling_units = missing_coordinates['polling_unit_name'].unique()

# Function to get coordinates from OpenCage API
def get_coordinates(address, api_key):
    url = f'https://api.opencagedata.com/geocode/v1/json?q={address}&key={api_key}'
    response = requests.get(url)
    if response.status_code == 200:
        results = response.json()['results']
        if results:
            return results[0]['geometry']['lat'], results[0]['geometry']['lng']
    return None, None

# Retry logic to handle potential connection issues
def get_coordinates_with_retry(address, api_key, retries=3, delay=2):
    for attempt in range(retries):
        try:
            lat, lng = get_coordinates(address, api_key)
            return lat, lng
        except requests.ConnectionError:
            if attempt < retries - 1:
                time.sleep(delay)
            else:
                return None, None

# Dictionary to store the results
coordinates_dict = {}

# Fetch coordinates for each unique polling unit with retry logic
for unit in unique_polling_units:
    lat, lng = get_coordinates_with_retry(unit + ', Akwa Ibom, Nigeria', api_key)
    coordinates_dict[unit] = (lat, lng)
    time.sleep(1)  # To avoid hitting the rate limit

# Update the original data with the fetched coordinates
for i, row in missing_coordinates.iterrows():
    polling_unit = row['polling_unit_name']
    lat, lng = coordinates_dict.get(polling_unit, (None, None))
    data.at[i, 'location.latitude'] = lat
    data.at[i, 'location.longitude'] = lng

# Save the updated data to a new CSV file
data.to_csv('updated_cleaned_data_no_duplicates.csv', index=False)
```

```python
from sklearn.neighbors import BallTree
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned data
cleaned_data = pd.read_csv('Akwa_Ibom_cleaned_data.csv')

# Function to convert degrees to radians
def deg2rad(deg):
    return deg * np.pi / 180

# Convert latitude and longitude to radians
cleaned_data['lat_rad'] = deg2rad(cleaned_data['location.latitude'])
cleaned_data['lon_rad'] = deg2rad(cleaned_data['location.longitude'])

# Create a BallTree for neighbor identification
coords = cleaned_data[['lat_rad', 'lon_rad']].values
tree = BallTree(coords, metric='haversine')

# Identify neighbors within a 1-kilometer radius (0.001 km in radians)
radius = 1 / 6371  # Earth's radius in kilometers
cleaned_data['neighbors'] = tree.query_radius(coords, r=radius).tolist()
```

```python
# Function to calculate average votes for each party
def calculate_average_votes(row, party):
    neighbor_indices = row['neighbors']
    if len(neighbor_indices) > 1:
        neighbor_indices = [i for i in neighbor_indices if i != row.name]  # Exclude the polling unit itself
    neighbor_votes = cleaned_data.loc[neighbor_indices, party]
    return neighbor_votes.mean()

parties = ['APC', 'LP', 'PDP', 'NNPP']
for party in parties:
    cleaned_data[f'{party}_avg'] = cleaned_data.apply(lambda row: calculate_average_votes(row, party), axis=1)

# Function to calculate outlier scores for each party
def calculate_outlier_scores(row, party):
    return abs(row[party] - row[f'{party}_avg'])

for party in parties:
    cleaned_data[f'{party}_outlier_score'] = cleaned_data.apply(lambda row: calculate_outlier_scores(row, party), axis=1)

# Save the data with outlier scores to a CSV file
cleaned_data.to_csv('Akwa_Ibom_outlier_scores.csv', index=False)
```

```python
# Load the cleaned data with outlier scores
outlier_scores = pd.read_csv('Akwa_Ibom_outlier_scores.csv')

# Sort by outlier score and select the top 3 outliers for each party
top_3_outliers = []
for party in parties:
    top_outliers = outlier_scores.nlargest(3, f'{party}_outlier_score')
    top_outliers['Party'] = party
    top_3_outliers.append(top_outliers)

top_3_outliers = pd.concat(top_3_outliers, ignore_index=True)

# Check column names to ensure correct plotting
print(top_3_outliers.columns)
```

```python
# Visualization: Bar chart of outlier scores for the top 3 outliers for each party
plt.figure(figsize=(12, 6))
sns.barplot(data=top_3_outliers, x='APC_outlier_score', y='polling_unit_name', hue='Party')
plt.title('Top 3 Outliers for Each Party by Outlier Score')
plt.xlabel('Outlier Score')
plt.ylabel('Polling Unit Name')
plt.legend(title='Party')
plt.tight_layout()
plt.show()
```
