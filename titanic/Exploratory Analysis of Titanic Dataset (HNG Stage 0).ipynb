```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic/train.csv')
```

```python
df.head()
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# Set the aesthetic style of the plots
sns.set(style="whitegrid")

#visualization showing survival rate
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()

#visualization showing survival rate by class
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival Count by Class')
plt.show()

#visualization showing survival rate by gender
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival Count by Gender')
plt.show()

# Plotting the distribution of Age for survivors and non-survivors
plt.figure(figsize=(10, 6))

# KDE plot
sns.kdeplot(df[df['Survived'] == 1]['Age'].dropna(), color="green", fill=True, label='Survived')
sns.kdeplot(df[df['Survived'] == 0]['Age'].dropna(), color="red", fill=True, label='Not Survived')

# Adding titles and labels
plt.title('Age Distribution of Titanic Passengers by Survival Status')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.show()
```
