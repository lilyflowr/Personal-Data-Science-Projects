```python
import pandas as pd
import requests
import time

# Load the CSV file
file_path = 'cleaned_data_no_duplicates.csv'
data = pd.read_csv(file_path)

# OpenCage API key
api_key = '83f25aa5c56e4d64b0ee5ee8658ae9d7'

# Extract unique polling unit names with missing coordinates
missing_coordinates = data[data['location.latitude'].isna() | data['location.longitude'].isna()]
unique_polling_units = missing_coordinates['polling_unit_name'].unique()

# Function to get coordinates from OpenCage API
def get_coordinates(address, api_key):
    url = f'https://api.opencagedata.com/geocode/v1/json?q={address}&key={api_key}'
    response = requests.get(url)
    if response.status_code == 200:
        results = response.json()['results']
        if results:
            return results[0]['geometry']['lat'], results[0]['geometry']['lng']
    return None, None

# Retry logic to handle potential connection issues
def get_coordinates_with_retry(address, api_key, retries=3, delay=2):
    for attempt in range(retries):
        try:
            lat, lng = get_coordinates(address, api_key)
            return lat, lng
        except requests.ConnectionError:
            if attempt < retries - 1:
                time.sleep(delay)
            else:
                return None, None

# Dictionary to store the results
coordinates_dict = {}

# Fetch coordinates for each unique polling unit with retry logic
for unit in unique_polling_units:
    lat, lng = get_coordinates_with_retry(unit + ', Akwa Ibom, Nigeria', api_key)
    coordinates_dict[unit] = (lat, lng)
    time.sleep(1)  # To avoid hitting the rate limit

# Update the original data with the fetched coordinates
for i, row in missing_coordinates.iterrows():
    polling_unit = row['polling_unit_name']
    lat, lng = coordinates_dict.get(polling_unit, (None, None))
    data.at[i, 'location.latitude'] = lat
    data.at[i, 'location.longitude'] = lng

# Save the updated data to a new CSV file
data.to_csv('updated_cleaned_data_no_duplicates.csv', index=False)
```

```python
# Create the voter turnout ratio feature
cleaned_data['voter_turnout_ratio'] = cleaned_data['Accredited_Voters'] / cleaned_data['Registered_Voters']

# Checking the new feature
print(cleaned_data[['polling_unit_name', 'ward_name', 'LGA', 'voter_turnout_ratio']].head())
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = cleaned_data['voter_turnout_ratio'].quantile(0.25)
Q3 = cleaned_data['voter_turnout_ratio'].quantile(0.75)
IQR = Q3 - Q1

# Define the outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = cleaned_data[(cleaned_data['voter_turnout_ratio'] < lower_bound) | (cleaned_data['voter_turnout_ratio'] > upper_bound)]

# Check the identified outliers
print(outliers[['polling_unit_name', 'ward_name', 'LGA', 'voter_turnout_ratio']].head())
```

```python
import folium

# Initialize the map centered around Akwa Ibom State
m = folium.Map(location=[cleaned_data['location.latitude'].mean(), cleaned_data['location.longitude'].mean()], zoom_start=10)

# Add all polling units to the map
for _, row in cleaned_data.iterrows():
    folium.Marker(
        location=[row['location.latitude'], row['location.longitude']],
        popup=f"{row['polling_unit_name']} - Voter Turnout Ratio: {row['voter_turnout_ratio']:.2f}",
        icon=folium.Icon(color='blue')
    ).add_to(m)

# Highlight outliers in red
for _, row in outliers.iterrows():
    folium.Marker(
        location=[row['location.latitude'], row['location.longitude']],
        popup=f"{row['polling_unit_name']} - Voter Turnout Ratio: {row['voter_turnout_ratio']:.2f}",
        icon=folium.Icon(color='red')
    ).add_to(m)

# Save the map to an HTML file
m.save('polling_units_map_with_outliers.html')

print("Map with outliers saved to 'polling_units_map_with_outliers.html'.")
```
